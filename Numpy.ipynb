###Задача 2
*(2 балла)*

Создайте массив из случайных значений из стандартного нормального распределения размера 6 на 6. 

Разверните массив таким образом, чтобы его размер стал 36 на 1.
Найдите среднее значение элементов массива.

**Разрешается использовать только методы Numpy. (Циклы и условия запрещены)**.

**Решение:**

```
# создадим массив из случайных значений из стандартного нормального распределения размера 6 на 6
norm_array = np.random.normal(0, 2, size=(6, 6))
print(norm_array)
```

```
# изменим размер массива на размер 36 на 1
norm_array = norm_array.reshape(36, 1)
print(norm_array)
```

```
# найдем среднее значение элементов массива
norm_array.mean()
```

###Задача 3

Создайте массив из 20 любых случайных значений, используя метод Numpy.
Найдите значения производных для созданного массива

**Решение:**

```
# создадим массив из 20 любых значений
new_array = np.random.random(20)
# найдем матрицу производных при помощи функции numpy.diff()
diff_array = np.diff(new_array)
print(new_array)
print(diff_array)
```

###Задача 4
*(2 балла)*

Дано выражение

Y = A * X * C

Создайте три случайных массива A, C, X. Размерности должны позволять выполнить матричное умножение согласно формуле. На выходе должен получиться массив Y с размерностью (5, 18). (Остальные размерности возьмите на свое усмотрение).

**Решение:**

```
# создадим массив А, C, X
A = np.random.randint(0, 10, (5,8))
C = np.random.randint(0, 10, (8,12))
X = np.random.randint(0, 10, (12,18))
# перемножим матрицы
Y = A.dot(C).dot(X)
print(Y)
print(Y.shape)
```
