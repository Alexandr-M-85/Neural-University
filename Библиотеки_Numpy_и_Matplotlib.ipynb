### Задача 1

Напишите функцию умножения трёх матриц без использования встроенных средств `np.dot()`, **@**.

```python
# Ваше решение
# функция умножения трех матриц
def matrix_multiply(matrix1, matrix2, matrix3):
    # проверяем размерность матриц на возможность умножения
    if matrix1.shape[1] != matrix2.shape[0] or matrix2.shape[1] != matrix3.shape[0]:
        print('Неправильные размеры матриц!')
    # функция умножения двух матриц
    def multiply(matrix1, matrix2):
        # создаем пустой список для результатов умножения
        new_matrix = []
        # пройдемся по всем строкам первой матрицы
        for i in range(matrix1.shape[0]):
            # пройдемся по столбцам второй матрицы
            for j in range(matrix2.shape[1]):
                # добавляем в список сумму произведений строки первой матрицы на столбец второй матрицы
                new_matrix.append(sum(matrix1[i] * matrix2[:, j]))
        # результирующий список преобразуем в матрицу нужного размера
        new_matrix = np.array(new_matrix).reshape(matrix1.shape[0], matrix2.shape[1])
        return new_matrix
    
    # умножаем полученную матрицу на третью матрицу и возвращаем результат
    return multiply(multiply(matrix1, matrix2), matrix3)


# сгенерируем три матрицы
matrix1 = np.random.randint(-10, 11, (3, 7))
matrix2 = np.random.randint(-10, 11, (7, 2))
matrix3 = np.random.randint(-10, 11, (2, 8))
# выведем матрицы
print('Матрица 1:', matrix1, sep='\n')
print('Матрица 2:', matrix2, sep='\n')
print('Матрица 3:', matrix3, sep='\n')
# выведем результат умножения
print()
print(matrix_multiply(matrix1, matrix2, matrix3))
# для проверки выведем результат умножения через встроенные функции
print()
print('Проверка')
print(matrix1@matrix2@matrix3)
```

### Задача 2

Нужно вывести на печать матрицу размером N*N, заполненную числами натурального ряда, начиная с первого элемента и двигаясь по спирали:

---

В зависимости от введенных данных результат может выглядеть так:

image.png

```python
# Ваше решение
# введем натуральное число
n=int(input('Размерность квадратной матрицы:'))
if n > 1:
    # сформируем массив из нулей 
    matrix = np.zeros((n, n))
    # введем индекс натурального числа
    idx = 1
    # несколько раз будем проходится по строкам и столбцам матрицы
    for x in range(n):
        # пройдемся по верхней заполненной нулями строке матрицы
        for j in range(n - 1):
            # если строка заполнена нулями, пройдемся по каждому элементу в строке и заполним натуральным числом
            if matrix[x][j] == 0:
                matrix[x][j] = idx
                idx += 1
        # пройдемся по последнему заполненному нулями столбцу
        for i in range(n - 1):
            # и сверху вниз заполним числами
            if matrix[i][n-1-x] == 0:
                matrix[i][n-1-x] = idx
                idx += 1
        # пройдемся по нижней заполненной нулями строке
        for j in range(n - 1, 0, -1):
            # и справа налево заменим нули числами
            if matrix[n-1-x][j] == 0:
                matrix[n-1-x][j] = idx
                idx += 1
        # пройдемся по первому заполненному нулями столбцу
        for i in range(n-1, 0, -1):
            # и снизу вверх заменим нули на числа
            if matrix[i][x] == 0:
                matrix[i][x] = idx
                idx += 1
    # выведем на печать
    print(matrix)
else:
    print('Неверное число!')
```

### Задача 4

Создайте гистограмму, содержащую различные варианты текстур.

---

В зависимости от введенных данных результат может выглядеть так:

image.png

```python
# Ваше решение
# сформируем x и y
x = np.arange(1, 11)
y = 3
hatches = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']

# сформируем полотно
plt.figure(figsize=(14, 7))
for i in range(len(x)):
    plt.bar(x[i], y, hatch=hatches[i])

# Заголовок графика
plt.title('Различные изображения гистограмм')
# Надпись для оси x
plt.xlabel('X')
# Надпись для оси y
plt.ylabel('Y')
# Вывод графика
plt.show()
```
