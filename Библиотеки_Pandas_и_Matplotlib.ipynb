Дан следующий исходный график:

```python
import matplotlib.pyplot as plt           # Импорт модуля pyplot
import numpy as np                        # Импорт библиотеки numpy
x = np.linspace(0, 20, 101)               # Генерация последовательности из 101 числа с шагом в 0.2
y = np.cos(x)                             # Запись cos(x) в переменную y

plt.plot(x, y)                            # Построение графика x, y
plt.show()                                # Фиксация графика
```

Измените отображение графика таким образом, чтобы получился такой результат:

Graphic.jpg

- размер полотна: 12,4
- ширина линии: 5
- цвет линии: **4b1234**
- тип маркера: h
- размер маркера: 10
- цвет маркера: **ff00d5**
- периодичность маркера: 5
- легенда графика закреплена в правом нижнем углу: **cos(x)**

- подписи осей: **ось X** и **ось Y**
- отрисована сетка

```python
# Ваше решение
import matplotlib.pyplot as plt                                                                          # Импорт модуля pyplot
import numpy as np                                                                                       # Импорт библиотеки numpy
x = np.linspace(0, 20, 101)                                                                              # Генерация последовательности из 101 числа с шагом в 0.2
y = np.cos(x)                                                                                            # Запись cos(x) в переменную y

plt.figure(figsize=(12, 4))                                                                              # задаем размер полотна
plt.plot(x, y, color='#4b1234', lw=5, marker='h', ms=10, mfc='#ff00d5', markevery=5, label='cos(x)')     # Построение графика x, y
plt.title('cos(x)')                                                                                      # заголовок графика
plt.xlabel('ось Х')                                                                                      # подпись оси Х
plt.ylabel('ось Y')                                                                                      # подпись оси Y
plt.legend(loc='lower right')
plt.grid()                                                                                               # отрисовка сетки
plt.show()                                                                                               # Фиксация графика
```

### Задача 3

Создайте резервную копию датафрейма с помощью специального метода. Удалите из датафрейма все строки, где есть хотя бы один пропуск.

Затем сохраните в переменную все уникальные категории приложений (столбец **Category**). Используя цикл, вычислите количество приложений для каждой категории.

```python
# Ваше решение
# сделаем копию датафрейма
df2 = df.copy()
# удалим строки с пропусками
df2.dropna(axis=0, how='any', inplace=True)
# сохраняем в переменную все уникальные категории приложений
category = df2.Category.unique()
# выведем количество приложений для каждой категории
for cat in category:
    print(cat, '-', df2[df2.Category == cat].shape[0])
```
